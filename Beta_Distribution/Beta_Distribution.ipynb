# Beta Distribution

```python
from IPython.display import display, Math, Latex
```

The Beta distribution is a continuous probability distribution defined on the interval $[0, 1]$. It is widely used in Bayesian statistics, machine learning, and various fields of economics to model random variables that are constrained to lie between $0$ and $1$ (e.g., probabilities, proportions).

- The random variable $X$ lies within the interval $[0, 1]$.
- The Beta distribution is parameterized by two positive shape parameters, $\alpha > 0$ and $\beta > 0$, which control the shape of the distribution.

The Beta function $ B(\alpha, \beta) $, which appears in the normalization constant of the Beta distribution, is defined as:

```python
display(Math(r'B(\alpha, \beta) = \int_0^1 t^{\alpha - 1} (1 - t)^{\beta - 1} dt'))
```

## Relationship Between Beta and Gamma Distributions

The Beta distribution and the Gamma distribution are closely related through their mathematical definitions and properties. This relationship arises because the Beta function, which normalizes the Beta distribution, can be expressed in terms of the Gamma function.

### 1. Beta Function and Gamma Function

This Beta function can be expressed in terms of the Gamma function $ \Gamma(x) $ as:
$$
B(\alpha, \beta) = \frac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
$$

Thus, the Beta distribution is inherently linked to the Gamma function via its normalization constant.


### 2. Derivation of Beta Distribution from Gamma Distribution

The Beta distribution can be derived as the distribution of the ratio of two independent Gamma-distributed random variables. Specifically:

- Let $ X_1 \sim \text{Gamma}(\alpha, 1) $ and $ X_2 \sim \text{Gamma}(\beta, 1) $ be two independent random variables following Gamma distributions with shape parameters $ \alpha $ and $ \beta $, respectively, and a common rate parameter of 1.
- Define a new random variable $ Y $ as the ratio:
$$
Y = \frac{X_1}{X_1 + X_2}
$$

Then, $ Y $ follows a Beta distribution with parameters $ \alpha $ and $ \beta $:
$$
Y \sim \text{Beta}(\alpha, \beta)
$$

This result shows that the Beta distribution naturally arises when considering normalized sums of Gamma-distributed random variables.


### 3. Intuition Behind the Relationship

- The Gamma distribution models the waiting time until the occurrence of a certain number of events in a Poisson process.
- The Beta distribution, on the other hand, models probabilities or proportions constrained to the interval $[0, 1]$.
- The connection between the two distributions reflects how proportions (Beta) can emerge from relative contributions of Gamma-distributed quantities.


### 4. Applications of the Relationship

The relationship between the Beta and Gamma distributions has practical applications in various fields:
- **Bayesian Statistics**: In Bayesian inference, the Beta distribution is often used as a conjugate prior for binomial likelihoods, while the Gamma distribution serves as a conjugate prior for Poisson likelihoods.
- **Reliability Analysis**: The Beta distribution can model failure probabilities, while the Gamma distribution models time-to-failure data.
- **Machine Learning**: The relationship is useful in probabilistic modeling, such as in topic modeling (e.g., Latent Dirichlet Allocation), where the Dirichlet distribution (a multivariate generalization of the Beta distribution) is derived from Gamma-distributed components.


### Summary

The Beta distribution is deeply connected to the Gamma distribution through the Beta function and the ratio of Gamma-distributed random variables. This relationship highlights the versatility of these distributions in modeling phenomena involving probabilities, proportions, and event times.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from IPython.display import Markdown, display

# Display a brief introduction
display(Markdown("### Simulation of Beta Distribution"))

# Parameters for the Beta distribution
alpha = 2  # Shape parameter alpha
beta_param = 5  # Shape parameter beta

# Generate random samples from the Beta distribution
np.random.seed(42)  # For reproducibility
samples = beta.rvs(alpha, beta_param, size=10000)

# Plot histogram of the samples
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='blue', label='Simulated Samples')

# Overlay the theoretical PDF
x = np.linspace(0, 1, 500)  # x values for the PDF
pdf_values = beta.pdf(x, alpha, beta_param)  # Compute the PDF
plt.plot(x, pdf_values, 'r-', lw=2, label=f'Theoretical PDF ($\\alpha={alpha}, \\beta={beta_param}$)')

# Add labels and legend
plt.title('Beta Distribution: Simulated Samples vs Theoretical PDF', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()
```

## Example 1

We will simulate how consumers allocate their budgets between two goods (e.g., "Good A" and "Good B") based on their preferences, which are modeled using the Beta distribution.

In microeconomics, consumer preferences often determine how they allocate their income across different goods. Suppose that consumers' preferences for "Good A" relative to "Good B" can be represented by a proportion p, where $p[0,1]$. Here:
- $p=0$: The consumer spends all their budget on Good B.
- $p=1$: The consumer spends all their budget on Good A.
- Intermediate values of p represent a mix of spending on both goods.

The Beta distribution is ideal for modeling these proportions because it is defined on the interval $[0,1]$ and can capture various shapes of preference distributions (e.g., uniform, skewed, or peaked around certain values). We will simulate the following scenario:

- Consumers have a fixed budget of $100.
- Their preferences for Good A are modeled using a Beta distribution with parameters  and .

Based on their preferences, each consumer allocates their budget between Good A and Good B:

- Amount spent on Good $A = p100$
- Amount spent on Good $B = (1p)100$

We will analyze the distribution of spending across consumers and interpret the results in terms of market demand.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from IPython.display import Markdown, display

# Display a brief introduction
display(Markdown("### Simulation of Consumer Preferences Using Beta Distribution"))

# Parameters for the Beta distribution
alpha = 4  # Shape parameter alpha
beta_param = 2  # Shape parameter beta

# Number of consumers
num_consumers = 10000

# Fixed budget per consumer
budget = 100

# Generate random samples of preferences (proportions) from the Beta distribution
np.random.seed(42)  # For reproducibility
preferences = beta.rvs(alpha, beta_param, size=num_consumers)

# Calculate spending on Good A and Good B
spending_good_a = preferences * budget
spending_good_b = (1 - preferences) * budget

# Plot histogram of spending on Good A
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(spending_good_a, bins=50, density=True, alpha=0.7, color='blue', label='Spending on Good A')
plt.title('Distribution of Spending on Good A', fontsize=14)
plt.xlabel('Amount Spent ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)

# Plot histogram of spending on Good B
plt.subplot(1, 2, 2)
plt.hist(spending_good_b, bins=50, density=True, alpha=0.7, color='green', label='Spending on Good B')
plt.title('Distribution of Spending on Good B', fontsize=14)
plt.xlabel('Amount Spent ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()

# Display summary statistics
display(Markdown(f"### Summary Statistics"))
display(Markdown(f"- Mean spending on Good A: ${np.mean(spending_good_a):.2f}"))
display(Markdown(f"- Mean spending on Good B: ${np.mean(spending_good_b):.2f}"))
display(Markdown(f"- Total spending across all consumers: ${np.sum(spending_good_a + spending_good_b):.2f}"))
```

#### Consumer Preferences:
- The Beta distribution allows us to model diverse consumer preferences. In this case, the parameters $=4$ and $=2$ suggest that consumers generally prefer Good A over Good B.

#### Market Demand :
- The simulation reveals the aggregate demand for each good. Firms producing Good A might observe higher sales compared to those producing Good B, given the skewed preferences.

#### Policy Implications :
- If policymakers want to encourage spending on Good B, they could introduce subsidies or incentives to shift consumer preferences.

## Example 2

We will model the distribution of household incomes in a developing country using the Beta distribution . This is a common approach in development economics because income distributions in such countries often exhibit significant inequality and are bounded within a finite range.

Income Distribution in Developing Countries

In many developing countries:
- Household incomes are often concentrated at lower levels, with a small fraction of households earning significantly higher incomes.
- The Beta distribution can be used to model income proportions (e.g., normalized incomes between 0 and 1) because it is defined on the interval $[0,1]$ and can capture skewed distributions.

We will simulate the following scenario:
- A population of households in a developing country.
- Household incomes are modeled as proportions of the maximum income in the country, normalized to lie between $0$ and $1$.
- The Beta distribution parameters ($$ and $$) reflect the degree of inequality in the population.

Using this simulation, we will analyze:
- The distribution of household incomes.
- Measures of inequality, such as the Gini coefficient .
- Policy implications for poverty reduction.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from IPython.display import Markdown, display

# Display a brief introduction
display(Markdown("### Simulation of Household Income Distribution in Development Economics"))

# Parameters for the Beta distribution
alpha = 2  # Shape parameter alpha (lower values increase skewness towards lower incomes)
beta_param = 5  # Shape parameter beta (higher values increase skewness towards lower incomes)

# Number of households
num_households = 10000

# Generate random samples of normalized incomes from the Beta distribution
np.random.seed(42)  # For reproducibility
normalized_incomes = beta.rvs(alpha, beta_param, size=num_households)

# Assume the maximum income in the country is $10,000
max_income = 10000
household_incomes = normalized_incomes * max_income

# Plot histogram of household incomes
plt.figure(figsize=(10, 6))
plt.hist(household_incomes, bins=50, density=True, alpha=0.7, color='orange', label='Household Incomes')
plt.title('Distribution of Household Incomes', fontsize=14)
plt.xlabel('Income ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Calculate summary statistics
mean_income = np.mean(household_incomes)
median_income = np.median(household_incomes)
poverty_line = 0.2 * max_income  # Poverty line set at 20% of maximum income
poverty_rate = np.mean(household_incomes < poverty_line)

# Calculate Gini coefficient (measure of income inequality)
sorted_incomes = np.sort(household_incomes)
n = len(sorted_incomes)
gini_coefficient = (2 * np.sum((np.arange(1, n + 1) * sorted_incomes)) / (n * np.sum(sorted_incomes))) - (n + 1) / n

# Display summary statistics
display(Markdown(f"### Summary Statistics"))
display(Markdown(f"- Mean income: ${mean_income:.2f}"))
display(Markdown(f"- Median income: ${median_income:.2f}"))
display(Markdown(f"- Poverty rate (below 20% of max income): {poverty_rate * 100:.2f}%"))
display(Markdown(f"- Gini coefficient (measure of inequality): {gini_coefficient:.2f}"))
```

#### Income Inequality :
- The skewed Beta distribution reflects the unequal distribution of incomes in developing countries, where a small fraction of households earn disproportionately high incomes.

#### Poverty :
- The poverty rate highlights the prevalence of low-income households, which is a critical issue in development economics.

#### Policy Implications :
Governments and policymakers can use this simulation to design targeted interventions, such as:
- Progressive taxation to redistribute income.
- Cash transfer programs to lift households above the poverty line.
- Investments in education and healthcare to improve long-term income prospects.

## Example 3

Beta distribution is used to model the daily proportion of active users on a digital platform over time. This is a common application in time series analysis for platforms like social media, e-commerce, or streaming services.

Suppose we are analyzing the proportion of active users on a digital platform over a 30-day period. The proportion of active users on any given day is constrained between 0 and 1, making the Beta distribution a natural choice for modeling this data. We will simulate:
- A time series of daily proportions of active users.
- Trends or seasonality in the data by varying the Beta distribution parameters ($$ and $$) over time.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from IPython.display import Markdown, display

# Display a brief introduction
display(Markdown("### Simulation of Time Series Using Beta Distribution"))

# Parameters for the Beta distribution
days = 30  # Number of days in the time series
beta_param = 5  # Fixed beta parameter

# Generate random samples for the increasing trend scenario
np.random.seed(42)  # For reproducibility
alpha_trend = np.linspace(2, 5, days)  # Increasing trend in alpha
daily_proportions_trend = [beta.rvs(alpha, beta_param) for alpha in alpha_trend]

# Generate random samples for the sinusoidal seasonality scenario
alpha_seasonal = 2 + 2 * np.sin(np.linspace(0, 2 * np.pi, days))  # Sinusoidal seasonality
daily_proportions_seasonal = [beta.rvs(alpha, beta_param) for alpha in alpha_seasonal]

# Plot the time series for the increasing trend scenario
plt.figure(figsize=(10, 6))
plt.plot(range(1, days + 1), daily_proportions_trend, marker='o', color='blue', label='Increasing Trend')
plt.title('Time Series: Increasing Trend in Proportion of Active Users', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Proportion of Active Users', fontsize=12)
plt.ylim(0, 1)  # Proportions are bounded between 0 and 1
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Plot the time series for the sinusoidal seasonality scenario
plt.figure(figsize=(10, 6))
plt.plot(range(1, days + 1), daily_proportions_seasonal, marker='o', color='green', label='Sinusoidal Seasonality')
plt.title('Time Series: Sinusoidal Seasonality in Proportion of Active Users', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Proportion of Active Users', fontsize=12)
plt.ylim(0, 1)  # Proportions are bounded between 0 and 1
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

# Calculate summary statistics for both scenarios
mean_proportion_trend = np.mean(daily_proportions_trend)
max_proportion_trend = np.max(daily_proportions_trend)
min_proportion_trend = np.min(daily_proportions_trend)

mean_proportion_seasonal = np.mean(daily_proportions_seasonal)
max_proportion_seasonal = np.max(daily_proportions_seasonal)
min_proportion_seasonal = np.min(daily_proportions_seasonal)

# Display summary statistics
display(Markdown(f"### Summary Statistics"))
display(Markdown(f"#### Increasing Trend Scenario"))
display(Markdown(f"- Mean proportion of active users: {mean_proportion_trend:.2f}"))
display(Markdown(f"- Maximum proportion of active users: {max_proportion_trend:.2f}"))
display(Markdown(f"- Minimum proportion of active users: {min_proportion_trend:.2f}"))

display(Markdown(f"#### Sinusoidal Seasonality Scenario"))
display(Markdown(f"- Mean proportion of active users: {mean_proportion_seasonal:.2f}"))
display(Markdown(f"- Maximum proportion of active users: {max_proportion_seasonal:.2f}"))
display(Markdown(f"- Minimum proportion of active users: {min_proportion_seasonal:.2f}"))
```

#### Increasing Trend :
- This scenario could represent a platform experiencing steady growth in user engagement due to marketing campaigns, feature improvements, or network effects.

#### Sinusoidal Seasonality :
- This scenario could represent periodic fluctuations in user engagement, such as higher activity during weekends, holidays, or specific events.

#### Comparison :
- The increasing trend suggests long-term growth, while the sinusoidal seasonality highlights short-term variability. Policymakers or platform managers might use these insights to design strategies that address both long-term trends and short-term fluctuations.
