# Negative Binomial Distribution

The Negative Binomial Distribution is a probabilistic model that describes the number of successes before a specified number of failures occurs in a sequence of independent Bernoulli trials, where each trial has the same probability of success. In economics, it is widely applied to model count data that exhibit overdispersion‚Äîwhere the variance exceeds the mean. Examples include the number of claims filed by an individual in insurance economics, the frequency of healthcare visits in health economics, or the number of defaults in credit risk modeling. This distribution's flexibility, stemming from its two parameters (the number of failures ùëü and the probability of success ùëù), allows it to capture varying degrees of skewness and kurtosis, making it an essential tool for analyzing economic phenomena involving discrete data. Its ability to accommodate overdispersion is particularly useful when traditional Poisson models fail, offering a robust alternative for real-world economic data analysis.

Assumptions:

Trials are independent.

Each trial has the same probability of success ($ p $).

The process continues until a fixed number of failures ($ r $) occur.

## Properties 

```python
# Probability Mass Function (PMF)
from IPython.display import Markdown, display
display(Markdown(r'<div align="left"> $P(X = x) = \binom{x + r - 1}{x} \cdot (1-p)^r \cdot p^x, \text{ where } r > 0, 0 < p \leq 1$ </div>'))
```

Mean: $ r(1-p)/p $ 

Variance: $ r(1-p)/p^2 $

Skewness and Kurtosis depend on the parameters $ r $ and $ p $.

$ r $ : Number of failures until the experiment is stopped (positive real number).

$ p $ : Probability of success in a single trial $ (0 < p <= 1) $.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Define the parameters for the Negative Binomial Distribution
r = 5  # Number of failures
p = 0.3  # Probability of success

# Generating a range of values for the random variable
x = np.arange(0, 50)

# Calculate the PMF using scipy's nbinom.pmf
pmf = nbinom.pmf(x, r, p)

# Plotting the Negative Binomial PMF
plt.figure(figsize=(10, 6))
plt.bar(x, pmf, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Negative Binomial Distribution (r=5, p=0.3)', fontsize=14)
plt.xlabel('Number of successes (x)', fontsize=12)
plt.ylabel('Probability Mass Function (PMF)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Additional Notes:
# 1. When p is small, the distribution becomes more spread out.
# 2. As r increases, the distribution approximates a normal distribution.
# 3. Negative Binomial Distribution is often used in modeling count data with overdispersion.
```

## Example 1

In microeconomics, firms might face uncertain demand for their products. Suppose we are studying the number of units sold for a product with highly variable demand, such as seasonal items or niche goods. The demand follows a Negative Binomial Distribution due to overdispersion (variance greater than the mean) caused by external shocks like economic conditions or consumer preferences.

We will simulate a scenario where firms experience random demand based on the Negative Binomial Distribution, considering the following parameters:

$ ùëü $: Expected number of failures before observing a certain level of success (captures the overdispersion of demand).

$ ùëù $: Probability of a single success (represents the likelihood of meeting demand in one attempt).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters for the Negative Binomial Distribution
r = 5  # Overdispersion parameter
p = 0.4  # Probability of success (likelihood of meeting demand)

# Simulating demand for 100 firms
num_firms = 100
demand = nbinom.rvs(r, p, size=num_firms)

# Visualizing the simulated demand
plt.figure(figsize=(10, 6))
plt.hist(demand, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Simulated Product Demand (Negative Binomial Distribution)', fontsize=14)
plt.xlabel('Units Sold', fontsize=12)
plt.ylabel('Frequency of Firms', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Microeconomic Analysis
print(f"Average demand: {np.mean(demand):.2f}")
print(f"Demand variance: {np.var(demand):.2f}")
print(f"Minimum demand: {np.min(demand)}")
print(f"Maximum demand: {np.max(demand)}")
```

In microeconomics, the Negative Binomial Distribution can model overdispersed demand for products, which is common in industries with seasonal or unpredictable consumption patterns. For instance, a firm producing niche goods might experience varying levels of product demand influenced by economic conditions or consumer preferences. By simulating demand across multiple firms using the Negative Binomial Distribution, we can observe how external shocks impact market behavior. The overdispersion parameter ($ùëü$) reflects variability, while the probability of success ($ùëù$) captures the likelihood of meeting demand under normal conditions. The simulation demonstrates that firms face significant differences in demand, ranging from extremely low to very high levels.

This variability has several implications for firms and policymakers. Firms must adapt their inventory and pricing strategies to accommodate uncertain demand, potentially holding higher stock levels to mitigate stockouts. On the other hand, excessive inventory might lead to increased holding costs if demand falls short. Policymakers can use insights from such simulations to design interventions, such as subsidies or support mechanisms, for industries facing volatile demand patterns. By understanding the distribution of demand, stakeholders can better allocate resources, stabilize markets, and support firms in managing risks associated with uncertainty.

## Example 2

In macroeconomics, the Negative Binomial Distribution can model the frequency of economic shocks that lead to recessions or periods of financial distress. Suppose we are studying the number of quarters until a country experiences a severe economic downturn (e.g., GDP contraction), where the likelihood of a downturn in any given quarter is low but varies due to overdispersion caused by global factors, policy responses, or market inefficiencies.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters for the Negative Binomial Distribution
r = 3  # Overdispersion parameter (resilience to economic shocks)
p = 0.2  # Probability of an economic downturn in a given quarter

# Simulating time to economic downturn for 50 countries
num_countries = 50
time_to_downturn = nbinom.rvs(r, p, size=num_countries)

# Visualizing the simulated time to economic downturn
plt.figure(figsize=(10, 6))
plt.hist(time_to_downturn, bins=15, color='lightcoral', edgecolor='black', alpha=0.7)
plt.title('Simulated Time to Economic Downturn (Negative Binomial Distribution)', fontsize=14)
plt.xlabel('Quarters Until Downturn', fontsize=12)
plt.ylabel('Frequency of Countries', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Macroeconomic Analysis
print(f"Average time to downturn: {np.mean(time_to_downturn):.2f} quarters")
print(f"Variance in time to downturn: {np.var(time_to_downturn):.2f} quarters^2")
print(f"Minimum time to downturn: {np.min(time_to_downturn)} quarters")
print(f"Maximum time to downturn: {np.max(time_to_downturn)} quarters")
```

In macroeconomics, the Negative Binomial Distribution can model the variability in the timing of economic downturns across countries. By simulating the number of quarters until a downturn, we observe how resilience ($ùëü$) and the probability of a shock ($ùëù$) affect the frequency and timing of recessions. The overdispersion parameter ($ùëü$) captures the degree to which countries differ in their ability to withstand shocks, reflecting differences in economic structures, policy frameworks, or external dependencies. Countries with high resilience experience fewer downturns over time, while those with low resilience face frequent challenges. The simulation highlights significant variability, with some countries showing prolonged stability and others quickly succumbing to adverse conditions.

These insights have practical implications for policymakers and global financial institutions. For example, countries with shorter times between downturns may require stronger fiscal buffers, counter-cyclical policies, or access to international safety nets like IMF support. On the other hand, countries with higher resilience can focus on long-term growth strategies. The simulation also underscores the importance of addressing global economic interdependencies, as shocks in one region can reduce resilience elsewhere. By understanding this variability, international organizations can better allocate resources, prepare for crises, and design tailored interventions to mitigate the impact of economic downturns on vulnerable populations.

## Example 3

In time series analysis, the Negative Binomial Distribution can model count data with overdispersion, such as the number of customer complaints, stock market transactions, or disease outbreaks over time. Suppose we are studying the number of complaints received by a utility company each day, where the variability in complaints is influenced by seasonal factors, service quality, and external events.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# Parameters for the Negative Binomial Distribution
r = 5  # Overdispersion parameter (resilience to high complaint days)
p = 0.3  # Probability of receiving no complaints in a given trial

# Simulating daily complaints over a 100-day period
days = 100
complaints = nbinom.rvs(r, p, size=days)

# Creating a time series plot
plt.figure(figsize=(12, 6))
plt.plot(range(1, days + 1), complaints, marker='o', linestyle='-', color='blue', alpha=0.7)
plt.title('Simulated Daily Complaints (Negative Binomial Distribution)', fontsize=14)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Number of Complaints', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Summary statistics
print(f"Average daily complaints: {np.mean(complaints):.2f}")
print(f"Variance in complaints: {np.var(complaints):.2f}")
print(f"Maximum complaints in a day: {np.max(complaints)}")
print(f"Minimum complaints in a day: {np.min(complaints)}")
```

Parameters:

$ùëü$: Reflects resilience to high complaint days. Higher ùëü indicates fewer days with excessive complaints.

$ùëù$: Probability of no complaints (or a low number of complaints) on a given trial, capturing the overall service quality.

Simulation:

A 100-day time series is generated using the Negative Binomial Distribution to simulate daily complaints received by a utility company. Each day's count is a random draw from the distribution, capturing overdispersion in complaint frequency.

Results:

The time series plot shows fluctuations in daily complaints, with most days experiencing moderate counts but occasional spikes due to poor service days or external factors.
Summary statistics provide insights into the average and variability of complaints over the observed period.
