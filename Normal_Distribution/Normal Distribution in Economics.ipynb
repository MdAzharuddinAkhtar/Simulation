The **Normal Distribution** is a foundational concept in statistics and econometrics. It is a continuous probability distribution characterized by its bell-shaped curve, which is symmetric about its mean. This distribution is defined by two parameters

1. **Mean \(μ)**: The central value around which the data is distributed.
2. **Standard Deviation \(σ)**: A measure of the dispersion or spread of the data around the mean.

```python
# The probability density function (PDF) for the normal distribution is given by:
from IPython.display import Markdown, display
display(Markdown(r'<div align="center"> $f(x|\mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2}$</div>'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate points for x-axis
x = np.linspace(-4, 4, 100)

# Calculate the normal distribution
y = norm.pdf(x, loc=0, scale=1)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', lw=2, label='Standard Normal Distribution')
plt.fill_between(x, y, alpha=0.3)

# Add labels and title
plt.title('Normal Distribution', fontsize=14)
plt.xlabel('X', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# Show the plot
plt.show()
```

```python
# change the parameters of the distribution
y = norm.pdf(x, loc=2, scale=1.5)
```

The normal distribution plays a crucial role in economics due to its natural occurrence in many aggregate phenomena, making it a cornerstone of statistical and econometric modeling. One fascinating fact is the Central Limit Theorem (CLT), which explains why the normal distribution is so prevalent in economic analysis. The CLT states that when independent random variables are summed, their normalized sum approaches a normal distribution, regardless of the original variables' distributions. This property is why metrics like GDP growth rates, inflation, and returns on investments often resemble a normal distribution in real-world data. Additionally, economists frequently assume normality when modeling errors in regression analysis, as it simplifies estimation and inference. While real-world data might exhibit skewness or heavy tails, the normal distribution's mathematical properties provide a robust foundation for developing models, testing hypotheses, and making predictions in various economic contexts.

### Key Properties

1. **Symmetry**: The distribution is symmetric around the mean.
2. **Empirical Rule**: Approximately 68% of the data lies within 1 standard deviation of the mean, 95% within 2, and 99.7% within 3

### Applications in Economics

#### Example 1: Income Distribution

Many economic phenomena, such as income distributions (in log form), returns on assets, and measurement errors, can often be approximated using a normal distribution.
Simulating Household Income Growth: In this simulation, we assume that the annual percentage growth in household income follows a normal distribution. We will generate data for 1,000 households over 10 years, assuming a mean annual growth rate of 5% and a standard deviation of 2%.

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting random seed for reproducibility
np.random.seed(42)

# Parameters
n_households = 1000  # Number of households
n_years = 10         # Number of years
mean_growth = 5      # Mean annual growth rate (%)
std_dev_growth = 2   # Standard deviation of growth rate (%)

# Simulating annual growth rates for each household
growth_rates = np.random.normal(loc=mean_growth, scale=std_dev_growth, size=(n_households, n_years))

# Assuming initial income of $50,000 for each household
initial_income = 50000
income_data = np.zeros((n_households, n_years))

# Calculating income over time
for i in range(n_households):
    income = initial_income
    for t in range(n_years):
        income *= (1 + growth_rates[i, t] / 100)  # Adjust for percentage
        income_data[i, t] = income

# Converting to a DataFrame for better visualization
income_df = pd.DataFrame(income_data, columns=[f"Year_{t+1}" for t in range(n_years)])

# Plotting income distribution for the first and last year
plt.figure(figsize=(12, 6))
sns.kdeplot(income_df["Year_1"], label="Year 1", shade=True)
sns.kdeplot(income_df[f"Year_{n_years}"], label=f"Year {n_years}", shade=True)
plt.title("Household Income Distribution Over Time")
plt.xlabel("Income")
plt.ylabel("Density")
plt.legend()
plt.show()
```

The graph illustrates the evolution of household income distribution from Year 1 to Year 10, showcasing the effects of compounded growth rates. Initially, in Year 1, incomes are tightly clustered around the starting value of $50,000, with a narrow spread and high density near the mean. However, by Year 10, the distribution shifts rightward, indicating higher overall incomes, and becomes noticeably wider, reflecting increased variation among households. This widening suggests that some households experienced faster growth rates over time, leading to greater income inequality. The compounding effects of even small differences in annual growth rates amplify disparities, creating a broader distribution in later years. This pattern aligns with real-world economic phenomena where heterogeneous growth opportunities contribute to long-term income stratification. Policymakers can leverage such insights to model and address inequality by designing interventions that reduce variability in income growth, ensuring more equitable economic progress.

```python
# Summary Statistics
print("Summary Statistics for Year 1")
print(income_df["Year_1"].describe())

print("\nSummary Statistics for Year 10")
print(income_df[f"Year_{n_years}"].describe())
```

#### Example 2: Income Distribution and Inequality

```python
# Simulating income data
np.random.seed(42)  # For reproducibility
```

```python
# Parameters
n_urban = 1000
n_rural = 1000
```

```python
# Simulated incomes: log-normal distribution (often used for income data)
urban_income = np.random.lognormal(mean=10, sigma=0.5, size=n_urban)  # Urban incomes
rural_income = np.random.lognormal(mean=9, sigma=0.7, size=n_rural)   # Rural incomes
```

```python
# Combine data
all_income = np.concatenate([urban_income, rural_income])
labels = ['Urban'] * n_urban + ['Rural'] * n_rural
```

```python
# Calculate Gini coefficient
def gini_coefficient(incomes):
    sorted_incomes = np.sort(incomes)
    n = len(incomes)
    cumulative_income = np.cumsum(sorted_incomes)
    relative_income = cumulative_income / cumulative_income[-1]
    gini = 1 - (2 / n) * np.sum(relative_income) + (1 / n)
    return gini
```

```python
# Gini coefficients for each group
gini_urban = gini_coefficient(urban_income)
gini_rural = gini_coefficient(rural_income)
gini_all = gini_coefficient(all_income)
```

```python
# Plot income distribution
plt.hist(urban_income, bins=50, alpha=0.7, label=f'Urban (Gini: {gini_urban:.2f})', color='blue')
plt.hist(rural_income, bins=50, alpha=0.7, label=f'Rural (Gini: {gini_rural:.2f})', color='green')
plt.title(f'Income Distribution (Overall Gini: {gini_all:.2f})')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

```python
# Print results
print(f"Gini Coefficient (Urban): {gini_urban:.2f}")
print(f"Gini Coefficient (Rural): {gini_rural:.2f}")
print(f"Gini Coefficient (Overall): {gini_all:.2f}")
```

The simulated income distributions for urban and rural populations reveal significant disparities in income levels and inequality. Urban incomes are centered around higher values with a narrower spread, indicating relatively higher and more uniform earnings, while rural incomes are lower on average and exhibit greater variability. The Gini coefficients quantify these disparities, with urban areas showing moderate inequality and rural areas experiencing higher inequality, reflecting limited economic opportunities or structural challenges in rural regions. The overall Gini coefficient combines both distributions, capturing inter- and intra-group disparities. These results highlight the persistent urban-rural divide in economic opportunities and underscore the need for targeted interventions, such as rural employment programs and equitable resource distribution, to reduce inequality and bridge the gap. The Gini coefficient serves as a crucial metric for policymakers to assess and address income inequality effectively.

#### Example 3: Time Series of Inflation Rates

```python
# Parameters for the normal distribution
np.random.seed(42)
n_years = 50  # Number of years
mean_inflation = 3.0  # Average inflation rate (%)
std_dev_inflation = 1.5  # Standard deviation of inflation rate (%)
```

```python
# Simulate inflation rates
inflation_rates = np.random.normal(loc=mean_inflation, scale=std_dev_inflation, size=n_years)
```

```python
# Create a time series
years = np.arange(1970, 1970 + n_years)
inflation_df = pd.DataFrame({'Year': years, 'Inflation Rate (%)': inflation_rates})
```

```python
# Plot the inflation rates over time
plt.figure(figsize=(10, 6))
plt.plot(years, inflation_rates, marker='o', linestyle='-', label='Simulated Inflation Rate')
plt.axhline(mean_inflation, color='red', linestyle='--', label=f'Average Inflation ({mean_inflation}%)')
plt.title('Simulated Inflation Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.grid()
plt.legend()
plt.show()
```

The graph illustrates the variation in inflation rates over a simulated 50-year period, ranging from 1970 to 2019. The blue line shows the yearly inflation rates, which fluctuate around the red dashed line, representing the mean inflation rate of 3.0%. The random nature of the fluctuations captures the inherent variability of inflation due to economic factors such as demand and supply shocks, policy interventions, or external events. Some years show inflation rates significantly above or below the average, highlighting periods of economic instability or unexpected shocks. The steady red line provides a benchmark to assess deviations, emphasizing the stability of the long-term mean despite short-term variations.

```python
# Print summary statistics
print("Summary Statistics:")
print(inflation_df.describe())
```

```python
# Check if the data follows a normal distribution (optional visualization)
plt.figure(figsize=(8, 5))
plt.hist(inflation_rates, bins=15, alpha=0.7, color='skyblue', edgecolor='black', density=True)
plt.title('Distribution of Simulated Inflation Rates')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Density')
plt.grid()
plt.show()
```

The graph presents a histogram of the inflation rates, showing their distribution over the 50-year period. The bell-shaped curve reflects the characteristics of a normal distribution, consistent with the assumption used in the simulation. Most inflation rates cluster around the mean of 3.0%, with fewer occurrences at the extreme high or low ends. This distribution suggests that inflation rates are generally predictable within a certain range, with occasional outliers. The density normalization provides a clear view of the relative frequency, reinforcing the idea that inflation rates typically exhibit moderate variation around the average.
