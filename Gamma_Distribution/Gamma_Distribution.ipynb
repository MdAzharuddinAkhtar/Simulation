# Gamma Distribution

The Gamma distribution is a two-parameter family of continuous probability distributions that is widely used in economics, finance, and other fields. It is particularly useful for modeling non-negative random variables, such as income, waiting times, or durations.

The Gamma distribution is defined by two parameters:

Shape parameter $(k>0)$ : Controls the shape of the distribution.
Scale parameter $(θ>0)$ : Controls the scale (spread) of the distribution.
Alternatively, the Gamma distribution can also be parameterized using:

Rate parameter $(β=1/θ)$ : The inverse of the scale parameter.

```python
from IPython.display import Markdown, display

properties = r'''
<div align="center"> 
$$ \text{PDF: } f(x; k, \theta) = \begin{cases} \frac{x^{k-1} e^{-x/\theta}}{\theta^k \Gamma(k)}, & x \geq 0 \\ 0, & x < 0 \end{cases} $$ 
</div>
<div align="center"> 
$$ \text{CDF: } F(x; k, \theta) = \begin{cases} \frac{\gamma(k, x/\theta)}{\Gamma(k)}, & x \geq 0 \\ 0, & x < 0 \end{cases} $$ 
</div>
<div align="center"> 
$$ \text{Expected Value: } E(X) = k \cdot \theta $$ 
</div>
<div align="center"> 
$$ \text{Variance: } \text{Var}(X) = k \cdot \theta^2 $$ 
</div>
<div align="center"> 
$$ \text{Skewness: } \frac{2}{\sqrt{k}} $$ 
</div>
<div align="center"> 
$$ \text{Kurtosis (Excess): } \frac{6}{k} $$ 
</div>
<div align="center"> 
$$ \text{Moments: } E[X^n] = \theta^n \cdot \frac{\Gamma(k + n)}{\Gamma(k)} $$ 
</div>
'''
display(Markdown(properties))
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Define parameters
k = 3  # Shape parameter
theta = 2  # Scale parameter

# Mean and Variance
mean = k * theta
variance = k * theta**2

# Skewness and Kurtosis
skewness = 2 / np.sqrt(k)
kurtosis = 6 / k

# Print properties
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")

# Generate random samples from Gamma distribution
np.random.seed(42)
samples = np.random.gamma(shape=k, scale=theta, size=1000)

# Plot histogram of samples
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Simulated Data')

# Plot PDF
x = np.linspace(0, 20, 1000)
pdf = stats.gamma.pdf(x, a=k, scale=theta)
plt.plot(x, pdf, 'r-', label='Gamma PDF')

# Add labels and legend
plt.title('Gamma Distribution Simulation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()

# Compute CDF
cdf_value = stats.gamma.cdf(5, a=k, scale=theta)
print(f"CDF at x=5: {cdf_value}")
```

## Example 1

Suppose we are modeling the time it takes for consumers to make a purchase decision after being exposed to an advertisement. This duration can be modeled using a Gamma distribution because:

The time is non-negative.
The distribution allows flexibility in modeling different shapes of waiting times.

Assume the average time to make a decision is $6$ days $(k⋅θ=6)$.
Set $k=3$ and $θ=2$ to satisfy the mean condition.
Simulate $1,000$ observations of decision times.
Analyze the simulated data to estimate the probability of making a decision within a certain time frame (e.g., $5$ days).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display

# Parameters for the Gamma distribution
k = 3  # Shape parameter
theta = 2  # Scale parameter

# Simulate decision times for 1000 consumers
np.random.seed(42)  # For reproducibility
decision_times = np.random.gamma(shape=k, scale=theta, size=1000)

# Analyze the simulated data
mean_simulated = np.mean(decision_times)
variance_simulated = np.var(decision_times)
skewness_simulated = stats.skew(decision_times)
kurtosis_simulated = stats.kurtosis(decision_times)

# Print summary statistics
print(f"Simulated Mean: {mean_simulated:.2f}")
print(f"Simulated Variance: {variance_simulated:.2f}")
print(f"Simulated Skewness: {skewness_simulated:.2f}")
print(f"Simulated Kurtosis: {kurtosis_simulated:.2f}")

# Plot histogram of simulated decision times
plt.figure(figsize=(10, 6))
plt.hist(decision_times, bins=30, density=True, alpha=0.6, color='g', label='Simulated Data')

# Overlay the theoretical Gamma PDF
x = np.linspace(0, 20, 1000)
pdf = stats.gamma.pdf(x, a=k, scale=theta)
plt.plot(x, pdf, 'r-', label='Gamma PDF')

# Add labels and legend
plt.title('Gamma Distribution Simulation: Consumer Decision Times')
plt.xlabel('Decision Time (days)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the probability of making a decision within 5 days
probability_within_5_days = stats.gamma.cdf(5, a=k, scale=theta)
display(Markdown(f"**Probability of making a decision within 5 days:** {probability_within_5_days:.2f}"))
```

## Example 2

Suppose we are modeling the duration of unemployment spells for individuals in a labor market. Unemployment durations are often modeled using non-negative distributions, such as the Gamma distribution, because they naturally capture the variability and skewness observed in real-world data. What is the average duration of unemployment, and what proportion of individuals remain unemployed for more than 12 weeks?

Assumptions:
The unemployment duration follows a Gamma distribution.
Based on historical data, we estimate the shape parameter k=4 and scale parameter $θ=3$ (weeks). These parameters imply:
Mean unemployment duration: $E(X)=k⋅θ=12$ weeks.
Variance: $Var(X)=k⋅θ_2=36$ weeks

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display

# Parameters for the Gamma distribution
k = 4  # Shape parameter
theta = 3  # Scale parameter

# Simulate unemployment durations for 10,000 individuals
np.random.seed(42)  # For reproducibility
unemployment_durations = np.random.gamma(shape=k, scale=theta, size=10000)

# Analyze the simulated data
mean_simulated = np.mean(unemployment_durations)
variance_simulated = np.var(unemployment_durations)
proportion_long_unemployed = np.mean(unemployment_durations > 12)

# Print summary statistics
print(f"Simulated Mean Duration: {mean_simulated:.2f} weeks")
print(f"Simulated Variance: {variance_simulated:.2f}")
print(f"Proportion Unemployed > 12 Weeks: {proportion_long_unemployed:.2f}")

# Plot histogram of simulated unemployment durations
plt.figure(figsize=(10, 6))
plt.hist(unemployment_durations, bins=30, density=True, alpha=0.6, color='b', label='Simulated Data')

# Overlay the theoretical Gamma PDF
x = np.linspace(0, 30, 1000)
pdf = stats.gamma.pdf(x, a=k, scale=theta)
plt.plot(x, pdf, 'r-', label='Gamma PDF')

# Add labels and legend
plt.title('Gamma Distribution Simulation: Unemployment Durations')
plt.xlabel('Unemployment Duration (weeks)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the probability of remaining unemployed for more than 12 weeks
probability_long_unemployed = 1 - stats.gamma.cdf(12, a=k, scale=theta)
display(Markdown(f"**Probability of Remaining Unemployed > 12 Weeks:** {probability_long_unemployed:.2f}"))
```

About $41\%$ of individuals remain unemployed for more than 12 weeks. This indicates that a significant portion of the population experiences prolonged unemployment spells.

## Example 3

Suppose we are analyzing the annual capital expenditure (CapEx) of firms in a specific industry. Capital expenditures are often modeled using distributions like the Gamma distribution because they are non-negative and exhibit variability across firms.

What is the average annual CapEx for firms in the industry, and what proportion of firms spend more than $50 million annually on CapEx?

Assumptions:
The annual CapEx follows a Gamma distribution.
Based on historical data, we estimate the shape parameter $k=2$ and scale parameter $θ=10$ (in millions of dollars). These parameters imply:
Mean CapEx: $E(X)=k⋅θ=20$ million dollars.
Variance: $Var(X)=k⋅θ_2=200$ million dollars

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from IPython.display import Markdown, display

# Parameters for the Gamma distribution
k = 2  # Shape parameter
theta = 10  # Scale parameter (in millions of dollars)

# Simulate annual CapEx for 10,000 firms
np.random.seed(42)  # For reproducibility
capex = np.random.gamma(shape=k, scale=theta, size=10000)

# Analyze the simulated data
mean_simulated = np.mean(capex)
variance_simulated = np.var(capex)
proportion_high_capex = np.mean(capex > 50)

# Print summary statistics
print(f"Simulated Mean CapEx: {mean_simulated:.2f} million dollars")
print(f"Simulated Variance: {variance_simulated:.2f}")
print(f"Proportion of Firms with CapEx > 50 Million: {proportion_high_capex:.2f}")

# Plot histogram of simulated CapEx
plt.figure(figsize=(10, 6))
plt.hist(capex, bins=30, density=True, alpha=0.6, color='orange', label='Simulated Data')

# Overlay the theoretical Gamma PDF
x = np.linspace(0, 100, 1000)
pdf = stats.gamma.pdf(x, a=k, scale=theta)
plt.plot(x, pdf, 'b-', label='Gamma PDF')

# Add labels and legend
plt.title('Gamma Distribution Simulation: Annual Capital Expenditure')
plt.xlabel('Annual CapEx (Millions of Dollars)')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the probability of spending more than 50 million dollars
probability_high_capex = 1 - stats.gamma.cdf(50, a=k, scale=theta)
display(Markdown(f"**Probability of Spending > 50 Million:** {probability_high_capex:.2f}"))
```

Only about $5\%$ of firms spend more than $\$50$ million annually on CapEx. This indicates that high CapEx is relatively rare and concentrated among a small subset of firms.
