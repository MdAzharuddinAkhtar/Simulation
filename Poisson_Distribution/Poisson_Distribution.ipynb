# Poisson distribution

The Poisson distribution is a discrete probability distribution that models the number of events occurring within a fixed interval of time or space. It is widely used in economics to analyze rare or random events, such as the number of customer arrivals at a store, the frequency of financial crises, or the incidence of natural disasters. This paper outlines the distribution’s assumptions, properties, and provides a graphical representation.

## Asumption

The Poisson distribution relies on three key assumptions:

Events are Independent: The occurrence of one event does not affect the probability of another.

Constant Rate: The average rate $(λ)$ at which events occur is constant over time or space.

Rare Events: The probability of more than one event occurring in a very small interval is negligible.

These assumptions make the Poisson distribution suitable for modeling low-probability, high-impact events.

## Properties

```python
from IPython.display import Markdown, display

properties = r'''
<div align="center"> $$ \text{PMF: } P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} \quad \text{for } k = 0, 1, 2, \dots $$ </div>

<div align="center"> $$ \text{Expected Value: } E(X) = \lambda $$ </div>

<div align="center"> $$ \text{Variance: } \text{Var}(X) = \lambda $$ </div>

<div align="center"> $$ \text{Skewness: } \text{Skewness} = \frac{1}{\sqrt{\lambda}} $$ </div>
'''

display(Markdown(properties))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_values = [1, 4, 10]  # Different rates of occurrence
k_values = np.arange(0, 20)  # Range of event counts

# Plot Poisson PMF for different lambda values
plt.figure(figsize=(10, 6))
for lam in lambda_values:
    plt.plot(k_values, poisson.pmf(k_values, lam), 'o-', label=f'λ = {lam}')

plt.title('Poisson Distribution for Different λ Values')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability P(X = k)')
plt.legend()
plt.grid(True)
plt.show()
```

Interpretation of the Graph 

$λ=1$: The distribution is highly skewed, with the highest probability at $k=0$ and $k=1$.

$λ=4$: The distribution becomes more symmetric, peaking at $k=4$.

$λ=10$: The distribution approximates a normal distribution, centered at $k=10$.

As $λ$ increases, the Poisson distribution becomes less skewed and more bell-shaped, reflecting the Central Limit Theorem.

## Example 1

A retail store wants to model the number of customers arriving per hour to optimize staffing levels. Customer arrivals are assumed to follow a Poisson process with an average rate of $λ=10$ customers per hour.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_customers = 10  # Average rate of customer arrivals per hour
hours = 1000  # Simulate 1000 hours
customer_counts = np.random.poisson(lambda_customers, hours)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(customer_counts, bins=range(0, 21), density=True, alpha=0.7, color='blue', edgecolor='black')
plt.title('Simulated Customer Arrivals per Hour')
plt.xlabel('Number of Customers')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Results
print(f"Simulated Average Customers per Hour: {np.mean(customer_counts):.2f}")
```

The store can use this simulation to predict peak hours and allocate staff efficiently. For example, if the probability of more than 15 customers arriving in an hour is low, the store can reduce staffing during off-peak hours.

## Example 2

A central bank wants to model the number of financial crises occurring in a decade to assess systemic risk. Historical data suggests an average of λ=2 crises per decade.

```python
# Parameters
lambda_crises = 2  # Average rate of financial crises per decade
decades = 1000  # Simulate 1000 decades
crisis_counts = np.random.poisson(lambda_crises, decades)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(crisis_counts, bins=range(0, 6), density=True, alpha=0.7, color='red', edgecolor='black')
plt.title('Simulated Financial Crises per Decade')
plt.xlabel('Number of Crises')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Results
print(f"Simulated Probability of No Crises in a Decade: {np.mean(crisis_counts == 0):.2f}")
print(f"Simulated Probability of at Least One Crisis: {np.mean(crisis_counts >= 1):.2f}")
```

The central bank can use this simulation to estimate the likelihood of multiple crises occurring in a short period, informing policies such as capital buffer requirements or stress testing.

## Example 3

An insurance company models the number of claims filed per year for a specific policy. Historical data suggests an average of 
$λ=5$ claims per year.

```python
# Parameters
lambda_claims = 5  # Average rate of claims per year
years = 1000  # Simulate 1000 years
claim_counts = np.random.poisson(lambda_claims, years)

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(claim_counts, bins=range(0, 15), density=True, alpha=0.7, color='green', edgecolor='black')
plt.title('Simulated Insurance Claims per Year')
plt.xlabel('Number of Claims')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

# Results
print(f"Simulated Average Claims per Year: {np.mean(claim_counts):.2f}")
print(f"Simulated Probability of More Than 10 Claims: {np.mean(claim_counts > 10):.2f}")
```

The insurance company can use this simulation to estimate the risk of high claim volumes and set premiums accordingly. For example, if the probability of more than 10 claims in a year is low, the company can offer competitive pricing while maintaining profitability.
