# Uniform Distribution

The uniform distribution is one of the simplest probability distributions in statistics. It describes a situation where all outcomes in a given interval are equally likely.

Assumptions
Equally Likely Outcomes: Each point within the specified interval $[ùëé,ùëè]$ has an equal probability of occurrence.

Bounded Interval: The distribution is defined over a finite interval $[ùëé,ùëè]$.

Continuous Variable: The uniform distribution is defined for a continuous random variable (there's also a discrete version, but we focus on the continuous case here).

```python
from IPython.display import Markdown, display

properties = r'''
<div align="center"> 
$$ \text{PDF: } f(x) = \frac{1}{b-a} \quad \text{for } a \leq x \leq b $$ 
</div>
<div align="center"> 
$$ \text{CDF: } F(x) = \begin{cases} 0, & x < a \\ \frac{x-a}{b-a}, & a \leq x \leq b \\ 1, & x > b \end{cases} $$ 
</div>
<div align="center"> 
$$ \text{Expected Value: } E(X) = \frac{a + b}{2} $$ 
</div>
<div align="center"> 
$$ \text{Variance: } \text{Var}(X) = \frac{(b-a)^2}{12} $$ 
</div>
<div align="center"> 
$$ \text{Skewness: } 0 $$ 
</div>
<div align="center"> 
$$ \text{Kurtosis (Excess): } -1.2 $$ 
</div>
<div align="center"> 
$$ \text{Moments: } E[X^k] = \frac{b^{k+1} - a^{k+1}}{(k+1)(b-a)} $$ 
</div>
'''

display(Markdown(properties))
```

```python
from IPython.display import Markdown, display
import numpy as np
import matplotlib.pyplot as plt

# Visualization of Uniform Distribution
# Parameters of the uniform distribution
a, b = 2, 6

# Generate values for the PDF and CDF
x = np.linspace(1, 7, 1000)
pdf = np.where((x >= a) & (x <= b), 1 / (b - a), 0)
cdf = np.where(x < a, 0, np.where(x > b, 1, (x - a) / (b - a)))

# Plotting the PDF and CDF
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Plot PDF
ax[0].plot(x, pdf, label='PDF', color='blue')
ax[0].fill_between(x, pdf, alpha=0.3, color='blue')
ax[0].set_title('PDF of Uniform Distribution')
ax[0].set_xlabel('x')
ax[0].set_ylabel('f(x)')
ax[0].set_xlim(1, 7)
ax[0].set_ylim(0, 0.4)
ax[0].legend()

# Plot CDF
ax[1].plot(x, cdf, label='CDF', color='green')
ax[1].set_title('CDF of Uniform Distribution')
ax[1].set_xlabel('x')
ax[1].set_ylabel('F(x)')
ax[1].set_xlim(1, 7)
ax[1].set_ylim(0, 1.1)
ax[1].legend()

plt.tight_layout()
plt.show()
```

### Example 1

We can use the uniform distribution to simulate scenarios where outcomes are equally likely. For example, consider a market demand simulation where consumers randomly value a product within a price range $[ùëé,ùëè]$

Below, we simulate:

1. Consumer valuations drawn from a uniform distribution.
2. A decision to purchase based on a fixed market price.

```python
# Parameters
a_income, b_income = 1000, 5000  # Income range in dollars
price = 20  # Price of a good
demand_sensitivity = 0.01  # Sensitivity of demand to income

# Simulating consumer incomes
np.random.seed(42)  # For reproducibility
n_consumers = 1000  # Number of consumers
incomes = np.random.uniform(a_income, b_income, n_consumers)

# Calculate demand for each consumer
demands = demand_sensitivity * incomes / price

# Visualize the distribution of incomes and corresponding demands
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Histogram of incomes
ax[0].hist(incomes, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax[0].set_title('Distribution of Consumer Incomes')
ax[0].set_xlabel('Income ($)')
ax[0].set_ylabel('Frequency')

# Scatter plot of income vs. demand
ax[1].scatter(incomes, demands, alpha=0.6, color='green', edgecolor='black')
ax[1].set_title('Income vs. Demand')
ax[1].set_xlabel('Income ($)')
ax[1].set_ylabel('Demand (Units)')

plt.tight_layout()
plt.show()
```

# Example 2

An application of the uniform distribution could involve simulating the allocation of government spending across different sectors to analyze its potential impact on GDP growth or other economic indicators.

```python
# Parameters
n_sectors = 5
sectors = ['Education', 'Healthcare', 'Infrastructure', 'Defense', 'Social Welfare']
spending_total = 1000  # Total spending in billion dollars
multipliers = [1.5, 1.8, 2.0, 1.2, 1.6]  # GDP multipliers for each sector

# Uniform allocation of spending
np.random.seed(42)
spending_allocation = np.random.uniform(0, spending_total / n_sectors, n_sectors)
spending_allocation = spending_allocation / spending_allocation.sum() * spending_total

# Calculating GDP contribution
gdp_contribution = spending_allocation * multipliers

# Visualize the spending allocation and GDP contribution
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for spending allocation
ax[0].bar(sectors, spending_allocation, color='orange', alpha=0.7, edgecolor='black')
ax[0].set_title('Government Spending Allocation Across Sectors')
ax[0].set_ylabel('Spending (Billion $)')

# Bar plot for GDP contribution
ax[1].bar(sectors, gdp_contribution, color='purple', alpha=0.7, edgecolor='black')
ax[1].set_title('GDP Contribution by Sector')
ax[1].set_ylabel('GDP Contribution (Billion $)')

plt.tight_layout()
plt.show()
```

This simulation assumes that government spending is uniformly distributed across several sectors, such as education, healthcare, infrastructure, defense, and social welfare. The impact of this spending on GDP growth is calculated based on predefined multipliers for each sector.

### Example 3

Suppose we want to assess how different forecasting models perform over a given time horizon. Here‚Äôs a simulation example in Python, showing how uniformly distributed error terms might affect time series forecasts:

Context: Assume a time series model predicts economic variables, such as GDP growth. Forecast errors are uniformly distributed within a range, representing a bounded uncertainty.

Goal: Simulate forecasts with errors uniformly distributed and visualize their effect.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
time_horizon = 20  # Number of time periods
true_values = np.linspace(2, 5, time_horizon)  # Simulated true GDP growth
error_low, error_high = -0.5, 0.5  # Uniform error range

# Simulating forecast errors
np.random.seed(42)
errors = np.random.uniform(error_low, error_high, time_horizon)

# Generating forecasts
forecasts = true_values + errors

# Visualizing true values and forecasts
plt.figure(figsize=(10, 6))
plt.plot(range(1, time_horizon + 1), true_values, label='True Values', color='blue', marker='o')
plt.plot(range(1, time_horizon + 1), forecasts, label='Forecasts', color='orange', marker='x')
plt.fill_between(range(1, time_horizon + 1), 
                 true_values + error_low, 
                 true_values + error_high, 
                 color='gray', alpha=0.2, label='Error Range')

plt.title('Time Series Forecasts with Uniformly Distributed Errors')
plt.xlabel('Time Period')
plt.ylabel('GDP Growth (%)')
plt.legend()
plt.grid()
plt.show()
```

The uniform distribution is a versatile tool in time series analysis, especially for modeling bounded uncertainty in forecast errors or shocks. In the provided simulation, we applied the uniform distribution to generate forecast errors for GDP growth rates within a predefined range, simulating bounded unpredictability. By adding these errors to baseline GDP forecasts, the simulation creates a range of plausible outcomes. The visualization highlights this range, providing valuable insights into how uncertainties might influence forecasts. This approach is particularly useful for sensitivity analysis, enabling analysts to assess how robust predictions are to bounded randomness. Such applications are common in economic policymaking, where understanding and communicating uncertainty is critical for informed decision-making. The bounded nature of uniform distribution makes it a practical choice for scenarios where extreme deviations are unlikely, ensuring realistic modeling of forecast variability.
